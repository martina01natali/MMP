```python
import numpy as np
import pandas as pd
import sympy as sy
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def func(x):
    return .5*x**2-1
```

```python
func_v = np.vectorize(func)
```

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.figure()
plt.plot(x, func_v(x))
plt.vlines(0, -10, 50, colors='black')
plt.hlines(0, -10, 10, colors='black')
```

```python
def func1(x):
    return -2/3*np.power(x, 3/2)-np.power(x, 1/2)
```

```python
x = np.linspace(0, 10, 50)
func1_v = np.vectorize(func1)

plt.figure()
plt.plot(x, func1_v(x))
plt.vlines(0, -10, 50, colors='black')
plt.hlines(0, 0, 10, colors='black')
```
